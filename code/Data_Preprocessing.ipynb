## **Data - importing and preprocessing**

```
import pandas as pd
```

```
df1 = pd.read_csv("/content/cloned-repo/Data/CWC_WaterLevels/Hemavathi.csv")
df2 = pd.read_csv("/content/cloned-repo/Data/CWC_WaterLevels/Harangi.csv")
df3 = pd.read_csv("/content/cloned-repo/Data/CWC_WaterLevels/kabini.csv")
df4 = pd.read_csv("/content/cloned-repo/Data/CWC_WaterLevels/KRS.csv")

frames = [df1, df2, df3, df4]
df = pd.concat(frames)
df["FLOW_DATE"] = pd.to_datetime(df["FLOW_DATE"], format='%Y-%m-%d') # lots of entries missing
```

```
# last 10 characters of UNIQUE_KEY symbolize date
df1['DATE'] = df1['UNIQUE_KEY'].str[-10:]
df2['DATE'] = df2['UNIQUE_KEY'].str[-10:]
df3['DATE'] = df3['UNIQUE_KEY'].str[-10:]
df4['DATE'] = df4['UNIQUE_KEY'].str[-10:]
df['DATE'] = df['UNIQUE_KEY'].str[-10:]
```

```
df1["DATE"] = pd.to_datetime(df1["DATE"])
df2["DATE"] = pd.to_datetime(df2["DATE"])
df3["DATE"] = pd.to_datetime(df3["DATE"])
df4["DATE"] = pd.to_datetime(df4["DATE"])
df["DATE"] = pd.to_datetime(df["DATE"])
```

```
df1 = df1.sort_values(by="DATE")
df2 = df2.sort_values(by="DATE")
df3 = df3.sort_values(by="DATE")
df4 = df4.sort_values(by="DATE") 
df = df.sort_values(by="DATE")
```

```
# feet to metre
df["RES_LEVEL_METRE"] = df["RES_LEVEL_FT"]
df["RES_LEVEL_METRE"] = pd.to_numeric(df["RES_LEVEL_METRE"], errors='coerce')
df["RES_LEVEL_METRE"].astype(float)
df["RES_LEVEL_METRE"] *= float(0.3048)
```

1 thousand million cubic feet<br> = 1 thousand x 1 million x ((0.3048)^3)<br> = 28316846.592 cubic meters (SI base unit)<br>
1 Tmcft = 28316846.592 m3 = 0.028316846592 bcm

```
0.3048*0.3048*0.3048*1000*1000000*(1/1000000000)
```

```
# tmcft to bcm
df["PRESENT_STORAGE_BCM"] = df["PRESENT_STORAGE_TMC"]
df["PRESENT_STORAGE_BCM"] = pd.to_numeric(df["PRESENT_STORAGE_BCM"], errors='coerce')
df["PRESENT_STORAGE_BCM"].astype(float)
df["PRESENT_STORAGE_BCM"] *= float(0.028316846592)
```

# **cusec**

a unit of flow (especially of water) equal to one cubic foot per second.

```
# feet to metre
df4["RES_LEVEL_METRE"] = df4["RES_LEVEL_FT"]
df4["RES_LEVEL_METRE"] = pd.to_numeric(df4["RES_LEVEL_METRE"], errors='coerce')
df4["RES_LEVEL_METRE"].astype(float)
df4["RES_LEVEL_METRE"] *= float(0.3048)
# tmcft to bcm
df4["PRESENT_STORAGE_BCM"] = df4["PRESENT_STORAGE_TMC"]
df4["PRESENT_STORAGE_BCM"] = pd.to_numeric(df4["PRESENT_STORAGE_BCM"], errors='coerce')
df4["PRESENT_STORAGE_BCM"].astype(float)
df4["PRESENT_STORAGE_BCM"] *= float(0.028316846592)
# cusecs to BCM water volume change in a day
df4["INFLOW_VOLUME"] = df4["INFLOW_CUSECS"]
df4["INFLOW_VOLUME"] = pd.to_numeric(df4["INFLOW_VOLUME"], errors='coerce')
df4["INFLOW_VOLUME"].astype(float)
df4["INFLOW_VOLUME"] *= float(0.028316846592*24*60*60/1000000000)
df4["OUTFLOW_VOLUME"] = df4["OUTFLOW_CUECS"]
df4["OUTFLOW_VOLUME"] = pd.to_numeric(df4["OUTFLOW_VOLUME"], errors='coerce')
df4["OUTFLOW_VOLUME"].astype(float)
df4["OUTFLOW_VOLUME"] *= float(0.028316846592*24*60*60/1000000000)
```
